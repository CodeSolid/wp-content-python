```python
import pandas as pd
# Read the posts CSV from the web import
pd.set_option('display.max_rows', None)    
posts = pd.read_csv("all_posts.csv", parse_dates=True)
# posts
posts.loc[posts['categories'].isin(['Pandas', 'Jupyter'])]
```

```python
# Summarize by Month and Year
posts.index = pd.DatetimeIndex(posts.date)
group1 = posts.groupby([(posts.index.year), (posts.index.month)]).sum("words")
# group1.columns = ["year", "date", "word_count"]
group2 = posts.groupby([(posts.index.year), (posts.index.month)]).count()
merged = group1.merge(group2, left_index=True, right_index=True)
result = merged.loc[:, ["word_count_x", "title"]]
result.columns = ["word_count", "number_of_posts"]
result["average_post"] = result.word_count / result.number_of_posts
result
#group2
#group2
```

```python
stats = {"Word Count Sum": [result.word_count.sum()], 
         "Total Posts:  ": [result.number_of_posts.sum()], 
         "Average post length": [result.word_count.sum() / result.number_of_posts.sum()]}
summary = pd.DataFrame(stats)
summary
```

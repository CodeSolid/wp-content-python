```python
import pandas as pd
from pandas.tseries.offsets import DateOffset
df = pd.read_csv("users.csv", names=["Day", "Users"], header=None)
df["Date"] = pd.period_range(start="2021-11-21", end="2022-10-10")
df["Month"] = df["Date"].dt.month
users = df[["Date", "Users", "Month"]]
```

```python
#users["Avg"] = users["Users"].rolling(20,0).mean().round(1)
# users["Avg"] = users["Users"].expanding(min_periods=0).mean().round(1)
# users["Avg"] = users["Users"].ewm(span=30).mean().round(1)
```

```python
users["Avg"] = users["Users"].rolling(40,0).mean().round(1)
users["Avg2"] = users["Users"].ewm(span=30).mean().round(1)
ax = users[["Avg", "Avg2"]].plot(label="Users", figsize=(16,8))
```

```python
users["Avg"] = users["Users"].rolling(10,0).mean().round(1)
# users["Avg"] = users["Users"].expanding(min_periods=0).mean().round(1)
ax = users[["Avg"]].plot(label="Users", figsize=(20,10))
```
